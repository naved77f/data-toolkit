Q1. What is NumPy, and why is it widely used in PythonW
M NumPy is a library in Python for numerical computing, offering support for arrays, matrices, and a variety of
mathematical operations.


Q2. How does broadcasting work in NumPyW
M Broadcasting is a method that allows NumPy to perform operations on arrays of different shapes by aligning
them in a compatible manner. For example, when a smaller array's dimensions do not match a larger array,
NumPy implicitly expands the smaller array's dimensions to match the larger one without copying data. This
optimization enhances performance and reduces memory usage.


Q3. What is a Pandas DataFrameW
M A DataFrame is a two-dimensional labeled data structure in Pandas that can hold data of different types in
columns, making it suitable for tabular data analysis.


Q4. Explain the use of the groupby() method in Pandasd
M The groupby() method in Pandas is used for grouping data based on one or more keys. After grouping,
various aggregation operations such as mean, sum, or count can be performed. For example, grouping sales
data by region and then calculating the total sales per region is a common use case. It simplifies data
summarization and analysis.


Q5. Why is Seaborn preferred for statistical visualizationsW
M Seaborn simplifies creating statistical plots by providing a high-level interface and default themes, making
the plots more informative and visually appealing.


Q6. What are the differences between NumPy arrays and Python listsW
M NumPy arrays are faster and more memory-efficient than Python lists. They support vectorized operations,
which allow performing element-wise computations without explicit loops. Unlike Python lists, NumPy arrays
require elements of the same type and support multi-dimensional arrays for more complex numerical
operations.


Q7. What is a heatmap, and when should it be usedW
M A heatmap is a graphical representation of data where individual values are represented as colors. It is used
to visualize relationships or correlations between variables in a dataset.


Q8. What does the term “vectorized operation” mean in NumPyW
M Vectorized operations in NumPy refer to performing element-wise computations across entire arrays without
explicit Python loops. For example, adding two arrays element by element can be written as c = a + b. This
approach is faster because the computations are executed at the low-level C or Fortran level and avoid the
overhead of Python loops.


Q9. How does Matplotlib differ from PlotlyW
M Matplotlib is ideal for static, publication-quality plots, while Plotly excels in creating interactive, web-based
visualizations for dashboards and presentations.

# Using seaborn's **`Tips`** dataset

### Q16. Load the tips dataset and calculate the average tip percentage for each day of the week.

```python
# Import necessary libraries
import seaborn as sns
import pandas as pd

# Load the `tips` dataset
tips = sns.load_dataset('tips')

# Calculate the tip percentage
tips['tip_percentage'] = (tips['tip'] / tips['total_bill']) * 100

# Calculate the average tip percentage by day
average_tip_percentage = tips.groupby('day')['tip_percentage'].mean()
print("Average Tip Percentage by Day:")
print(average_tip_percentage)
```

### Q17. Create a bar plot comparing total bill amounts across different days of the week.

```python
# Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load the `tips` dataset
tips = sns.load_dataset('tips')

# Create a bar plot comparing total bill amounts across different days
sns.barplot(data=tips, x='day', y='total_bill', estimator='mean', ci=None)
plt.title('Average Total Bill Amount by Day')
plt.ylabel('Average Total Bill')
plt.xlabel('Day')
plt.show()
```

### Q18. Create a scatter plot of total_bill vs. tip to analyze the relationship between them.

```python
# Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load the `tips` dataset
tips = sns.load_dataset('tips')

# Create a scatter plot of total_bill vs. tip
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='time')
plt.title('Total Bill vs. Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

### Q19. Visualize the distribution of total bill amounts using a histogram.

```python
# Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load the `tips` dataset
tips = sns.load_dataset('tips')

# Visualize the distribution of total bill amounts using a histogram
sns.histplot(data=tips, x='total_bill', bins=15, kde=True)
plt.title('Distribution of Total Bill Amounts')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.show()
```

# Using Seaborn's **`Diamonds`** dataset

### Q21. Load the diamonds dataset and calculate the average price of diamonds for each cut category

```python
# Import necessary libraries
import seaborn as sns
import pandas as pd

# Load the `diamonds` dataset
diamonds = sns.load_dataset('diamonds')

# Calculate the average price of diamonds for each cut category
average_price_cut = diamonds.groupby('cut')['price'].mean()
print("Average Price of Diamonds by Cut:")
print(average_price_cut)
```

### Q22. Create a histogram to visualize the distribution of diamond carat weights.

```python
# Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load the `diamonds` dataset
diamonds = sns.load_dataset('diamonds')

# Create a histogram to visualize the distribution of diamond carat weights
sns.histplot(data=diamonds, x='carat', bins=20, kde=True)
plt.title('Distribution of Diamond Carat Weights')
plt.xlabel('Carat')
plt.ylabel('Frequency')
plt.show()
```

### Q23. Generate a scatter plot of price vs. carat, categorized by clarity.

```python
# Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load the `diamonds` dataset
diamonds = sns.load_dataset('diamonds')

# Generate a scatter plot of price vs. carat, categorized by clarity
sns.scatterplot(data=diamonds, x='carat', y='price', hue='clarity', alpha=0.7)
plt.title('Price vs. Carat Categorized by Clarity')
plt.xlabel('Carat')
plt.ylabel('Price')
plt.legend(title='Clarity', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

### Q24. Create a bar plot to compare the mean price of diamonds across different color categories.

```python
# Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load the `diamonds` dataset
diamonds = sns.load_dataset('diamonds')

# Create a bar plot to compare the mean price of diamonds across different color categories
sns.barplot(data=diamonds, x='color', y='price', estimator='mean', ci=None, palette='muted')
plt.title('Mean Price of Diamonds by Color')
plt.xlabel('Diamond Color')
plt.ylabel('Mean Price')
plt.show()
```

### Q25. Create a heatmap of the correlation between numerical features in the diamonds dataset.

```python
# Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load the `diamonds` dataset
diamonds = sns.load_dataset('diamonds')

# Create a heatmap of the correlation between numerical features in the diamonds dataset
correlation_matrix = diamonds.corr(numeric_only=True)
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Heatmap of Correlations in the Diamonds Dataset')
plt.show()
```

# Using Seaborn's **`penguins`** Dataset

### Q26. Load the Penguins dataset and calculate the average body mass for each species.

```python
# Import necessary libraries
import seaborn as sns
import pandas as pd

# Load the `penguins` dataset
penguins = sns.load_dataset('penguins')

# Calculate the average body mass for each species
average_body_mass_species = penguins.groupby('species')['body_mass_g'].mean()
print("Average Body Mass by Species (grams):")
print(average_body_mass_species)
```

### Q27. Create a violin plot to visualize the distribution of flipper lengths grouped by species.

```python
# Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load the `penguins` dataset
penguins = sns.load_dataset('penguins')

# Create a violin plot to visualize the distribution of flipper lengths grouped by species
sns.violinplot(data=penguins, x='species', y='flipper_length_mm', palette='Set3')
plt.title('Distribution of Flipper Lengths by Species')
plt.xlabel('Species')
plt.ylabel('Flipper Length (mm)')
plt.show()
```

### Q28. Generate a scatter plot of bill length vs. bill depth, categorized by species.

```python
# Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load the `penguins` dataset
penguins = sns.load_dataset('penguins')

# Generate a scatter plot of bill length vs. bill depth, categorized by species
sns.scatterplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', hue='species', style='species')
plt.title('Bill Length vs. Bill Depth Categorized by Species')
plt.xlabel('Bill Length (mm)')
plt.ylabel('Bill Depth (mm)')
plt.legend(title='Species')
plt.show()
```

### Q29. Create a box plot to compare the body mass of male and female penguins.

```python
# Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load the `penguins` dataset
penguins = sns.load_dataset('penguins')

# Create a box plot to compare the body mass of male and female penguins
sns.boxplot(data=penguins, x='sex', y='body_mass_g', hue='species', palette='coolwarm')
plt.title('Box Plot of Body Mass by Sex and Species')
plt.xlabel('Sex')
plt.ylabel('Body Mass (grams)')
plt.legend(title='Species')
plt.show()
```

### Q30. Create a heatmap of correlations between numerical features in the Penguins dataset.

```python
# Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load the `penguins` dataset
penguins = sns.load_dataset('penguins')

# Create a heatmap of correlations between numerical features in the Penguins dataset
correlation_matrix_penguins = penguins.corr(numeric_only=True)
sns.heatmap(correlation_matrix_penguins, annot=True, cmap='viridis', fmt='.2f')
plt.title('Heatmap of Correlations in the Penguins Dataset')
plt.show()
```
